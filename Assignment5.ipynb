```python
def factorial1(n):

    my_list = [x+1 for x in range(n)]

    fact = 1
    for x in my_list:
        fact *= x
    return fact

print(factorial1(5))
print(factorial1(10))
print(factorial1(20))
print(factorial1(50))
print(factorial1(100))
```

```python
def special_factorial1(n):
    fact = 1
    for x in range(1, n+1):
        if x != 4:
            fact *= x
    return fact

print(special_factorial1(5))
print(special_factorial1(10))
print(special_factorial1(20))
print(special_factorial1(50))
print(special_factorial1(100))
```
